# Proyek Analisis Data

## Menyiapkan semua library yang dibutuhkan

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Wrangling

### Gathering Data

```python
customers_df = pd.read_csv('data//customers_dataset.csv')
product_df = pd.read_csv('data//products_dataset.csv')
orders_df = pd.read_csv('data//orders_dataset.csv')
order_items_df = pd.read_csv('data//order_items_dataset.csv')
order_review_df = pd.read_csv('data//order_reviews_dataset.csv')
```

#### Menentukan kolom tabel yang dipakai

```python
customers_df = customers_df.loc[:, ['customer_id', 'customer_city']]
customers_df.head()
```

```python
product_df = product_df.loc[:, ['product_id', 'product_category_name']]
product_df.head()
```

```python
orders_df = orders_df.loc[:, ['order_id', 'customer_id','order_status','order_delivered_customer_date','order_estimated_delivery_date']]
orders_df.head()
```

```python
order_items_df = order_items_df.loc[:, ['order_id', 'product_id','price','freight_value']]
order_items_df.head()
```

```python
order_review_df = order_review_df.loc[:, ['order_id', 'review_score']]
order_review_df.head()
```

#### Menggabungkan data yang telah dipilih kolomnya

```python
orders_customers_df = pd.merge(
    left=orders_df,
    right=customers_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
orders_customers_df.head()
```

```python
orders_customers_df = pd.merge(
    left=orders_customers_df,
    right=order_review_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
orders_customers_df.head()
```

```python
product_order_df = pd.merge(
    left=order_items_df,
    right=product_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
product_order_df.head()
```

```python
main_data = pd.merge(
    left=orders_customers_df,
    right=product_order_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
main_data.head()
```

### Assessing Data

#### Menilai tabel `main_data`

```python
main_data.info()
```

Dengan memilih colom ` order_delivered_customer_date`, maka data yang diambil merupakan data dengan status `delivered` atau terkirim sehingga pada analisis ini dibatasi hanya status dengan customer yang sudah selesai melakukan transaksi 

### Cleaning Data

#### Membersihkan Tabel `main_data`

##### Missing value

```python
main_data.isna().sum()
```

Banyaknya missing value disebabkan oleh banyaknya proses pesanan yang belum selesai seperti status `approved`, `processing`,`shipped` dan lain-lain

Missing value dihapus karena hanya data yang memiliki `order_delivered_customer_date` saja yang akan diambil

```python
main_data = main_data.dropna()
```

```python
main_data.isna().sum()
```

Selanjutnya dilakukan penghapusan pada data yang terduplikasi untuk mengurangi bias yang terjadi

```python
print("Jumlah duplikasi: ", main_data.duplicated().sum())
```

```python
main_data.drop_duplicates(inplace=True)
```

```python
print("Jumlah duplikasi: ", main_data.duplicated().sum())
```

#### Memperbaiki tipe data

```python
main_data.info()
```

Pada data `order_delivered_customer_date` dan `order_estimated_delivery_date` masih bertipe object dan mengandung jam, maka perlu diubah menjadi `date`

```python
datetime_columns = ["order_delivered_customer_date", "order_estimated_delivery_date"]

for column in datetime_columns:
  main_data[column] = pd.to_datetime(main_data[column])

for column in datetime_columns:
    main_data[column] = main_data[column].dt.date

for column in datetime_columns:
  main_data[column] = pd.to_datetime(main_data[column])
```

Pada data `review_score` masih bertipe float sedangkan nilai review merupakan nilai integer

```python
main_data['review_score']= main_data['review_score'].astype(int)
```

Kemudian mengubah kolom lain menjadi string

```python
string_columns = ["order_id", "customer_id","order_status","customer_city","product_id","product_category_name"]

for column in string_columns:
  main_data[column] = main_data[column].astype(str)
```

```python
main_data.head()
```

## Explanatory Data Analysis (EDA)

### Explore `main_data`

```python
main_data.describe(include="all")
```

### Membuat kolom `Barang_tiba`

Pada data terdapat kolom `order_delivered_customer_date` dan `order_estimated_delivery_date` sehingga dapat dibuat kolom baru `Barang_tiba` yang memberikan informasi, bahwa barang datang terlambat atau tepat waktu 

```python
main_data['Barang_tiba'] = 'Tepat Waktu' 
main_data.loc[main_data['order_delivered_customer_date'] > main_data['order_estimated_delivery_date'], 'Barang_tiba'] = 'Terlambat' 
```

```python
main_data.head()
```

Dapat dicari hubungan antara barang terlambat dengan review yang diberikan

```python
main_data.groupby(by="Barang_tiba").agg({
    "customer_id": "nunique",
    "review_score": ["max", "min", "mean",]
})
```

Mengurutkan `customer_city` dengan jumlah barang terlambat paling banyak

```python
result = main_data.groupby(by="customer_city").agg({
    "customer_id": "nunique",
    "Barang_tiba": lambda x: (x == 'Terlambat').sum()
})

result_sorted = result.sort_values(by='Barang_tiba', ascending=False)

print(result_sorted)
```

Berdasarkan hasil tersebut menunjukkan jika Kota Sao Paulo merupakan kota dengan customer terbanyak tetapi diikuti dengan jumlah kedatangan yang terlambat paling banyak

Kemudian dapat dicari juga kota dengan rata-rata biaya pengiriman paling mahal

```python
result = main_data.groupby(by="customer_city").agg({
    "customer_id": "nunique",
    "freight_value": ["max", "min", "mean"]
})

# Mengurutkan hasil berdasarkan nilai freight_value tertinggi
result_sorted = result.sort_values(by=('freight_value', 'max'), ascending=False)

print(result_sorted)
```

Berdasarkan hasil tersebut, Kota Sao Raimundo Nonato merupakan kota dengan rata-rata biaya pengiriman termahal

Kemudian dapat dicari juga kategori produk yang paling laris atau paling banyak dibeli

```python
result = main_data.groupby(by="product_category_name").agg({
    "customer_id": "nunique",
    "review_score":  "mean",
    "price": "mean"
})

result_sorted = result.sort_values(by='customer_id', ascending=False)

print(result_sorted)
```

Kategori produk yang paling banyak dibeli adalah cama mesa banho atau bed bath table

Kemudian dapat dicari juga kategori produk yang membelikan pemasukan terbesar kepada perusahaan

```python
result = main_data.groupby(by="product_category_name").agg({
    "customer_id": "nunique",
    "review_score":  "mean",
    "price": "mean"
})

# Menambahkan kolom keuntungan
result['profit'] = result['customer_id'] * result['price']

# Mengurutkan DataFrame berdasarkan kolom keuntungan
result_sorted = result.sort_values(by='profit', ascending=False)

print(result_sorted)
```

Berdasarkan hasil tersebut, kategori produk yang paling banyak memberikan keuntungan adalah beleza saude atau health and beauty

Kemudian dapat dicari juga jumlah keuntungan pada setiap bulan dari tahun 2016 - 2018

```python
# Kelompokkan data berdasarkan bulan dan hitung jumlah pesanan dan total harga
monthly_orders_df = main_data.resample(rule='M', on='order_delivered_customer_date').agg({
    "order_id": "nunique",
    "price": "sum"
})

# Ubah indeks menjadi format 'Tahun-Bulan'
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%B')

# Mengganti nama kolom
monthly_orders_df.rename(columns={
    "order_id": "order_count",
    "price": "revenue"
}, inplace=True)

monthly_orders_df.reset_index(inplace=True)
print(monthly_orders_df)
```

Terlihat dari hasil bahwa keuntungan tersbesar didapatkan pada bulan Agustus tahun 2018

## Visualization & Explanatory Analysis

### Bagaimana performa penjualan dan revenue perusahaan dalam beberapa bulan terakhir?

```python
# Membuat plot garis
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_orders_df["order_delivered_customer_date"],
    monthly_orders_df["order_count"],
    marker='o', 
    linewidth=2,
    color="#72BCD4"
)
plt.title("Total Keuntungan", loc="center", fontsize=20)
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotasi label sumbu x sebesar 45 derajat dengan perataan ke kanan
plt.yticks(fontsize=10)
plt.show()
```

Berdasarkan hasil tersebut, keuntungan perusahaan terus menunjukkan angka yang baik hingga bulan Agustus tahun 2018. Terjadi penurunan yang tajam dikarenakan data customer pada bulan September dan Oktober tahun 2018 kebanyakan belum dalam status `delivered` sehingga belum terekap. Namun secara keseluruhan, dari beberapa periode sebelumnya menunjukkan peningkatan yang konstan dan positif

### Kategori produk apa yang paling banyak dan paling sedikit terjual?

```python
# Menghitung jumlah produk yang terjual untuk setiap produk
sum_order_items_df = main_data.groupby("product_category_name").size().reset_index(name="quantity_x")

# Membuat gambar dan sumbu
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

# Plot produk terbaik
sns.barplot(x="quantity_x", y="product_category_name", data=sum_order_items_df.sort_values(by="quantity_x", ascending=False).head(5), palette="Blues", ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Product Category", loc="center", fontsize=18)
ax[0].tick_params(axis='y', labelsize=15)

# Plot produk terburuk
sns.barplot(x="quantity_x", y="product_category_name", data=sum_order_items_df.sort_values(by="quantity_x", ascending=True).head(5), palette="Reds_r", ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product Category", loc="center", fontsize=18)
ax[1].tick_params(axis='y', labelsize=15)

# Menambahkan judul keseluruhan
plt.suptitle("Best and Worst Performing Product Category by Number of Sales", fontsize=20)
plt.show()
```

Berdasarkan hasil tersebut, terlihat 5 besar kategori produk paling banyak dan paling sedikit dibeli. Kategori produk paling banyak dibeli adalah cama mesa banho atau bed bath table, disusul beleza saude atau health beauty, esporte lazer atau sport leisure, informatica acessorios atau computer accessories, dan moveis decoracao atau furniture decor.


Sedangkan kategori produk paling sedikit dibeli dalah seguros e servicos atau security and service, disusul fashion roupa infanto juvenil atau fashion childrens clothes, pc gamer atau computer, cds dvds musicais atau kaset musik, dan la cuisine atau bahan masakan

### Apakah keterlambatan dalam pengiriman dapat mempengaruhi rating produk?

```python
average_review_by_status = main_data.groupby('Barang_tiba')['review_score'].mean()

# Plotting
plt.figure(figsize=(10, 6))

# Plot rata-rata review pada masing-masing status pengiriman
sns.barplot(x=average_review_by_status.index, y=average_review_by_status.values, palette='Set2')
plt.title('Rata-rata Review Berdasarkan Status Pengiriman', fontsize=16)
plt.xlabel('Status Pengiriman', fontsize=14)
plt.ylabel('Rata-rata Review', fontsize=14)

# Menampilkan nilai rata-rata review pada setiap batang
for i, val in enumerate(average_review_by_status.values):
    plt.text(i, val + 0.05, f'{val:.2f}', horizontalalignment='center', fontsize=12)

plt.show()
```

Berdasarkan hasil tersebut, terlihat bahwa pengiriman tepat waktu memiliki rata-rata review rating sebesar 4.26 yang mana nilai tersebut berbeda jauh dengan rata-rata rating pada pengiriman terlambat yaitu sebesar 2.27 sehingga berdasarkan hasil tersebut dapat disimpulkan jika pengiriman terlambat dapat mempengaruhi atau memberikan rating buruk pada produk.

### Kota manakah yang menyumbangkan customer paling banyak?

```python
# Menghitung jumlah pelanggan di setiap kota
customer_count_by_city = main_data['customer_city'].value_counts()

# Memilih lima kota teratas
top_5_cities = customer_count_by_city.head(5)

# Plotting
plt.figure(figsize=(10, 5))
colors_ = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
sns.barplot(
    x=top_5_cities.values, 
    y=top_5_cities.index,
    palette=colors_
)
plt.title("Top 5 Cities by Number of Customers", loc="center", fontsize=15)
plt.ylabel("City", fontsize=12)
plt.xlabel("Number of Customers", fontsize=12)
plt.xticks(fontsize=10)
plt.show()
```

Berdasarkan hasil tersebut, terlihat 5 besar kota dengan penyumbang customer paling banyak. Posisi pertama ditempati oleh Sao Paulo, diikuti oleh Rio De Janeiro, Belo Horizonte, Brasilia, dan Curitiba
